```python
import pandas as pd
import random
from IPython.display import display

#Esse código gera senhas aleatórias com base no arquivo excel fornecido e salva em um arquivo csv na pasta Downloads

def gerar_senha():
    numero = random.randint(10000, 10999)
    return str(numero)

df_tabela = pd.read_excel(r"", engine='xlrd', skiprows=2, header=None, skipfooter=4) #caminho do arquivo excel

df_tabela['Senha'] = df_tabela.apply(lambda _: gerar_senha(), axis=1)

df_resultado = df_tabela[[1, 'Senha']]

display(df_resultado)

df_resultado.to_csv(r"C:\Users\danie\Downloads\.csv", index=False, header=False) #salva o arquivo na pasta Downloads em csv

print("Senhas geradas e salvas na pasta Downloads com sucesso!")
```
